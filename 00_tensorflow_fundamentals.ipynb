##  Creating random tensors

Random tensors are tensors of some arbitrary size which contain random numbers.

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_2 = random_2.normal(shape=(3,2))
random_2
```

##  Getting information from tensors

when dealing with tensor you probably want to be aware of the following attributes:

* Shape
* Rank
* Axis or dimension
* Size

```python
# create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along axis 0 of tensor:", rank_4_tensor.shape[0])
print("Elements along last axis of tensor:", rank_4_tensor.shape[-1])
print("Total number of elements (2*3*4*5):", tf.size(rank_4_tensor).numpy())
```

###  Indexing tensors

```python
# get the first 2 elements of each dimensions
rank_4_tensor[:2,:2,:2,:2]
```

```python
# get each first element from each dimension from each index except for the final one
rank_4_tensor[:1,:1,:1,:]
```

```python
# Create a rank 2 tensor (2 dimensions)
rank_2_tensor = tf.constant([[10,7],
                             [3,4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
# Get the last item of each of row of our rank 2 tensor
rank_2_tensor[:,-1]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[..., tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1) # "-1" means expand the final axis
```

##  Aggregating Tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values.

```python
# get the absolute values
D = tf.constant([-7,-10])
D
```

```python
tf.abs(D)
```

**Let's go through the following forms of aggregation:**

* Get the minimum
* get the maximum
* get the mean of the tensor
* get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0,100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Finf the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# find the sum
tf.reduce_sum(E)
```

```python
# find the varanice, W tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# find the standard std
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

```python
# find the maximum value in the tensor
tf.reduce_max(E)
```

```python
# find the mimimum value in the tensor
tf.reduce_min(E)
```

### FInd the positional maximum and minimum

```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# FInd the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# find the maximum value in the tensor
tf.reduce_max(F)
```

```python
# Check if equal
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using th positional minimum index
F[tf.argmin(F)]
```
